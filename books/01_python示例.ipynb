## 1. 输出：让程序开口说话

Python 最常用的输出函数是 `print()`。你可以把它想象成程序的扩音器。

```python
print("你好，Python！")
```

## 2. 输入 + 输出：和用户打招呼

真实场景里，输入通常来自 `input()`。
为了让这个 Notebook 一打开就能看到效果，这里先用“模拟输入”的方式演示。

```python
# 假设这是用户在 input() 中输入的名字
user_name = "小林"
print(f"你好，{user_name}！欢迎来到 Python 世界。")
```

下面是同样逻辑的真实输入写法（运行后会等待你在输入框中输入内容）：

```python
user_name = input("请输入你的名字：")
print(f"你好，{user_name}！欢迎来到 Python 世界。")
```

## 3. 输入商品单价和数量，输出总价

重点提醒：`input()` 得到的是字符串（str）。
- 单价通常要转成浮点数 `float`
- 数量通常要转成整数 `int`

这样才能正确计算总价。

```python
# 模拟用户输入
price_str, count_str = "12.5", "3"

# 字符串 -> 数值
price = float(price_str)  # 商品单价
count = int(count_str)    # 商品数量

total = price * count

print("商品单价:", price)
print("商品数量:", count)
print("商品总价:", total)
```
