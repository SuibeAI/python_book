## Python 高级数据类型：列表与元组

- 理解：元组（tuple）适合表示结构固定、不希望随意修改的一组数据。
- 会使用：元组表示一笔商品销售记录 `(商品单价, 销售数量)`。

## 为什么需要列表和元组？

前面几章中，一个变量通常只保存一个值。例如：

```python
sales_day1 = 12
```

但真实问题经常需要保存“一组相关的值”。例如同一件商品连续 3 天的销量：第 1 天卖出 12 件，第 2 天卖出 15 件，第 3 天卖出 9 件。

如果每一天都单独写一个变量，数据一多就很难管理。列表和元组就是用来组织多个值的基础数据结构。

```{mermaid}
flowchart TD
    A[多个相关数据] --> B{数据是否会继续增删改?}
    B -- 会 --> C[使用列表 list]
    B -- 不会，结构固定 --> D[使用元组 tuple]
```

这张图只需要先建立直觉：列表更适合“可变化的一组数据”；元组更适合“结构固定的一条记录”。

## 列表 `list`

### 数据定义与范围

列表（list）可以把多个值按顺序放在一起。列表用方括号 `[]` 表示，里面的值用逗号隔开。

本节用列表表示同一件商品连续 3 天的销售数量：

```python
sales = [12, 15, 9]
```

它表示：

| 位置 | 含义 | 值 |
|---|---|---:|
| 第 1 个值 | 第 1 天销量 | `12` |
| 第 2 个值 | 第 2 天销量 | `15` |
| 第 3 个值 | 第 3 天销量 | `9` |

````python
sales = [12, 15, 9]

print("连续 3 天销量:", sales)
print("数据类型:", type(sales))
print("天数:", len(sales))
````

### 存储与表示规则

列表是“有顺序”的数据结构。Python 会给列表中的每个位置一个编号，这个编号叫索引（index）。

索引从 `0` 开始，不是从 `1` 开始：

| 表达式 | 含义 | 结果 |
|---|---|---:|
| `sales[0]` | 第 1 天销量 | `12` |
| `sales[1]` | 第 2 天销量 | `15` |
| `sales[2]` | 第 3 天销量 | `9` |

初学时最容易出错的是：第 1 个位置的索引是 `0`。

````python
sales = [12, 15, 9]

first_day = sales[0]
second_day = sales[1]
third_day = sales[2]

print("第 1 天销量:", first_day)
print("第 2 天销量:", second_day)
print("第 3 天销量:", third_day)
````

### 支持的运算

列表最常见的操作包括读取、修改、追加和遍历。本章先只使用最基础的几种。

| 操作 | 示例 | 说明 |
|---|---|---|
| 读取元素 | `sales[0]` | 读取第 1 天销量 |
| 修改元素 | `sales[2] = 10` | 修改第 3 天销量 |
| 追加元素 | `sales.append(11)` | 添加第 4 天销量 |
| 统计长度 | `len(sales)` | 得到列表中有几个值 |
| 遍历列表 | `for count in sales` | 依次取出每一天销量 |

下面先计算连续 3 天总销量。

````python
sales = [12, 15, 9]

total_sales = 0

for count in sales:
    total_sales = total_sales + count

print("连续 3 天销量:", sales)
print("总销量:", total_sales)
````

### 列表可以修改

列表适合保存会变化的数据。例如发现第 3 天销量录入错误，原来写成 `9`，实际应该是 `10`，就可以通过索引修改。

````python
sales = [12, 15, 9]

print("修改前:", sales)

sales[2] = 10

print("修改后:", sales)
print("第 3 天销量:", sales[2])
````

## 元组 `tuple`

### 数据定义与范围

元组（tuple）也可以把多个值放在一起。元组用圆括号 `()` 表示，里面的值用逗号隔开。

本节用元组表示一笔商品销售记录：

```python
record = (20.0, 3)
```

它表示：商品单价是 `20.0` 元，销售数量是 `3` 件。

| 位置 | 含义 | 值 |
|---|---|---:|
| 第 1 个值 | 商品单价 | `20.0` |
| 第 2 个值 | 销售数量 | `3` |

````python
record = (20.0, 3)

print("商品销售记录:", record)
print("数据类型:", type(record))
print("字段个数:", len(record))
````

### 存储与表示规则

元组也有顺序，也可以用索引读取。和列表一样，索引从 `0` 开始。

| 表达式 | 含义 | 结果 |
|---|---|---:|
| `record[0]` | 商品单价 | `20.0` |
| `record[1]` | 销售数量 | `3` |

区别在于：元组通常用来表示结构固定的数据。这里的第 1 个位置总是单价，第 2 个位置总是数量。

````python
record = (20.0, 3)

price = record[0]
count = record[1]
total = price * count

print("商品单价:", price)
print("销售数量:", count)
print("销售金额:", total)
````

### 支持的运算

元组支持读取、统计长度、拆包等操作。它不适合修改内部元素。

| 操作 | 示例 | 说明 |
|---|---|---|
| 读取元素 | `record[0]` | 读取商品单价 |
| 统计长度 | `len(record)` | 得到字段数量 |
| 拆包 | `price, count = record` | 把元组中的值分别交给变量 |

拆包适合结构清晰的元组。比如 `(商品单价, 销售数量)` 这类固定格式。

````python
record = (20.0, 3)

price, count = record
total = price * count

print("商品单价:", price)
print("销售数量:", count)
print("销售金额:", total)
````

## 列表和元组的区别

| 对比维度 | 列表 `list` | 元组 `tuple` |
|---|---|---|
| 写法 | `[12, 15, 9]` | `(20.0, 3)` |
| 典型含义 | 一组同类数据 | 一条结构固定的记录 |
| 是否常用于修改 | 常用于修改 | 通常不修改 |
| 本章例子 | 连续 3 天销量 | `(商品单价, 销售数量)` |

简单判断：如果你要保存“多天销量”这类同类数据，用列表更自然；如果你要保存“一笔销售记录”这类固定结构，用元组更自然。

## 复合结构：列表中保存元组

前面我们分别使用过列表和元组：

- 列表适合保存一组同类数据，例如连续 3 天销量。
- 元组适合保存一条结构固定的记录，例如 `(商品单价, 销售数量)`。

真实问题中，二者经常会组合使用。比如一天内卖出了多件商品，每件商品都可以写成一个元组：

```python
(商品单价, 销售数量)
```

多件商品放在一起，就可以写成一个列表：

```python
[(商品单价, 销售数量), (商品单价, 销售数量), ...]
```

这时，外层列表表示“多条记录”，内层元组表示“一条记录”。

````python
records = [(20.0, 3), (15.0, 2), (8.0, 5)]

total = 0

for record in records:
    price = record[0]
    count = record[1]
    amount = price * count
    total = total + amount
    print("一件商品销售金额:", amount)

print("当天销售总金额:", total)
````

这段代码的执行过程是一个 `for` 循环：

1. 程序先准备一个初始值 `total = 0`。
2. `for` 循环从 `records` 中取出一条商品记录 `record`。
3. 从元组 `record` 中取出单价和数量。
4. 计算这一件商品的销售金额，并累加到 `total` 中。
5. 如果 `records` 中还有下一条记录，就继续重复；如果没有，就输出当天销售总金额。

```{mermaid}
flowchart TD
    A["开始"] --> B["total = 0"]
    B --> C{"records 中还有记录吗"}
    C -- "有" --> D["取出下一条 record"]
    D --> E["price = record[0]"]
    D --> F["count = record[1]"]
    E --> G["amount = price * count"]
    F --> G
    G --> H["total = total + amount"]
    H --> C
    C -- "没有" --> I["输出 total"]
```

这张图要注意循环箭头：`total = total + amount` 执行完以后，程序会回到条件判断，继续查看 `records` 中是否还有下一条商品记录。

这里的关键不是“语法变复杂了”，而是数据层次变多了：列表负责保存多条记录，元组负责固定每条记录内部的含义。

## 嵌套列表：多天销售记录

如果要记录多天销售情况，可以继续向外增加一层列表。

例如：

- 第 1 天卖出两件商品。
- 第 2 天卖出两件商品。
- 第 3 天卖出两件商品。

每一天都是一个“商品销售记录列表”；多天记录再放进一个更大的列表中。

````python
day1 = [(20.0, 3), (15.0, 2)]
day2 = [(20.0, 4), (8.0, 5)]
day3 = [(15.0, 1), (8.0, 6)]

all_days = [day1, day2, day3]

all_total = 0

for day_records in all_days:
    day_total = 0

    for record in day_records:
        price = record[0]
        count = record[1]
        amount = price * count
        day_total = day_total + amount

    print("一天销售金额:", day_total)
    all_total = all_total + day_total

print("多天销售总金额:", all_total)
````

这段代码中出现了两层 `for` 循环：

- 外层循环：一次处理一天的销售记录。
- 内层循环：一次处理这一天中的一件商品。

```{mermaid}
flowchart TD
    A["开始"] --> B["all_total = 0"]
    B --> C{"all_days 中还有下一天吗"}
    C -- "有" --> D["取出一天 day_records"]
    D --> E["day_total = 0"]
    E --> F{"day_records 中还有商品吗"}
    F -- "有" --> G["取出一条 record"]
    G --> H["price = record[0]"]
    G --> I["count = record[1]"]
    H --> J["amount = price * count"]
    I --> J
    J --> K["day_total = day_total + amount"]
    K --> F
    F -- "没有" --> L["输出 day_total"]
    L --> M["all_total = all_total + day_total"]
    M --> C
    C -- "没有" --> N["输出 all_total"]
```

这张图中有两个回环：内层回环负责把某一天的每件商品都算完；外层回环负责把每一天都处理完。

可以把嵌套结构理解为“从外到内逐层打开”：先找到某一天，再找到这一天中的某件商品，最后取出单价和数量。
