## 循环结构：重复执行一类操作

上一章中，我们已经学习了顺序结构和选择结构。顺序结构解决“先做什么、后做什么”；选择结构解决“不同情况走哪条路”。

循环结构解决的是另一个问题：当一类操作需要重复执行多次时，如何避免把相似代码写很多遍？

例如购买同一件商品 3 件时，每一件都要计算价格。如果一行一行手写，会很重复。循环可以把“对每一件商品做同样的事”表达出来。

```{mermaid}
flowchart TD
    A[开始] --> B[准备一组数据]
    B --> C{还有下一个数据吗?}
    C -- 有 --> D[取出一个数据]
    D --> E[执行重复操作]
    E --> C
    C -- 没有 --> F[结束]
```

这张图要观察两点：第一，循环会回到条件判断处；第二，每次循环只处理当前这一个数据。

## `for` 循环

### 数据定义与适用场景

`for` 循环适合处理“数量已知”或“数据已经列出来”的情况。比如：

- 已知同一件商品买 3 件，需要逐件计算价格。
- 已知有几门课程成绩，需要逐个累加。
- 已知有几个字符，需要逐个查看编码。

本节继续使用商品购买例子。这里假设购买的是同一件商品，单价固定为 `unit_price`。为了表示第 1 件、第 2 件、第 3 件，先用一个简单列表保存商品件号。

### 执行流程

```{mermaid}
flowchart TD
    A[准备商品件号列表] --> B[total = 0]
    B --> C{还有下一件吗?}
    C -- 有 --> D[取出当前件号]
    D --> E[total = total + unit_price]
    E --> C
    C -- 没有 --> F[输出 total]
```

`for item_no in item_numbers` 可以读作：依次从 `item_numbers` 中取出一个件号，临时命名为 `item_no`，然后执行缩进块里的代码。

````python
unit_price = 100
item_numbers = [1, 2, 3]

total = 0

for item_no in item_numbers:
    total = total + unit_price

print("商品单价:", unit_price)
print("购买数量:", 3)
print("应付总价:", total)
````

> 这里先提前介绍一下列表（list）数据结构：它可以把多个值按顺序放在一起，例如 `item_numbers = [1, 2, 3]` 表示第 1 件、第 2 件、第 3 件商品。列表的完整用法会在后续章节详细介绍；本节只需要知道，`for` 可以从列表中依次取出每一个值。

### `for` 循环中的变量变化

阅读循环代码时，要关注两个变量：

- `item_no`：每次循环取出的当前商品件号。
- `total`：到目前为止已经累加出来的总价。

下面把每一步的变化打印出来。

````python
unit_price = 100
item_numbers = [1, 2, 3]

total = 0

print("开始 total:", total)

for item_no in item_numbers:
    print("当前第几件:", item_no)
    total = total + unit_price
    print("累加后 total:", total)

print("最终应付总价:", total)
````

## `for` 和 `if` 的组合

### 数据定义与适用场景

循环负责“逐件处理商品”，选择结构负责“判断当前这一件是否享受折扣”。两者组合后，就能表达更复杂的规则。

假设商场规则是：

- 第 1 件商品按原价计算。
- 从第 2 件商品开始，每件商品 6 折。

这里购买的是同一件商品，单价不变；变化的是第几件商品采用什么折扣。

### 执行流程

```{mermaid}
flowchart TD
    A[准备商品件号列表] --> B[total = 0]
    B --> C[取出当前件号]
    C --> D{当前是第 1 件吗?}
    D -- 是 --> E[按原价计算]
    D -- 否 --> F[按 6 折计算]
    E --> G[累加到 total]
    F --> G
    G --> H{还有下一件吗?}
    H -- 有 --> C
    H -- 没有 --> I[输出 total]
```

在这个流程里，循环和选择结构各自负责一件事：`for` 负责一件一件处理，`if` 负责判断当前件号用哪种价格规则。

````python
unit_price = 100
item_numbers = [1, 2, 3]

total = 0

for item_no in item_numbers:
    if item_no == 1:
        pay_price = unit_price
    else:
        pay_price = unit_price * 0.6

    total = total + pay_price

    print("第", item_no, "件原价:", unit_price)
    print("第", item_no, "件实付:", pay_price)

print("最终应付总价:", total)
````

## `for` 和 `while` 的区别

### 数据定义与适用场景

`for` 和 `while` 都能表达重复执行，但适用场景不同。

| 结构 | 更适合的场景 | 判断依据 | 商品例子 |
|---|---|---|---|
| `for` | 已经知道要处理哪些数据 | 数据集合是否处理完 | 已知同一商品买 3 件，逐件累加价格 |
| `while` | 不确定要重复多少次 | 某个条件是否仍成立 | 预算还够，就继续购买下一件同款商品 |

简单判断：如果你已经有一组数据要逐个处理，优先用 `for`；如果你只知道“满足某个条件就继续”，可以考虑 `while`。

### `while` 的执行流程

```{mermaid}
flowchart TD
    A[开始] --> B[设置初始条件]
    B --> C{条件是否成立?}
    C -- 是 --> D[执行循环体]
    D --> E[更新条件相关变量]
    E --> C
    C -- 否 --> F[结束]
```

`while` 循环最容易出错的地方是忘记更新条件相关变量。如果条件一直成立，循环就会一直执行下去。

````python
budget = 200
price = 60
count = 0

while budget >= price:
    budget = budget - price
    count = count + 1

print("每件商品价格:", price)
print("最多可购买数量:", count)
print("剩余预算:", budget)
````

### 1. 概念判断：为什么需要循环？

如果同一件商品购买 3 件、30 件、300 件都要计算总价，为什么不建议一行一行手写加法？

<details>
<summary>参考答案</summary>

因为每一件商品的处理逻辑相似，手写很多行会重复、容易漏改，也不容易扩展。循环可以把“对每一件商品做同样的计算”表达成统一结构。

</details>

### 2. 概念判断：`for` 和 `while` 怎么选？

已知同一件商品要买 3 件，应该优先使用 `for` 还是 `while`？如果规则是“预算够就继续买”，更适合哪个？

<details>
<summary>参考答案</summary>

已知要处理第 1、2、3 件商品时，优先使用 `for`，因为数据已经列出来了。规则是“预算够就继续买”时，更适合 `while`，因为循环次数取决于预算条件何时不成立。

</details>

### 3. 概念判断：`for + if` 中两种结构各自负责什么？

在“同一商品第二件开始 6 折”的例子中，`for` 和 `if` 分别负责什么？

<details>
<summary>参考答案</summary>

`for` 负责逐件处理商品；`if` 负责判断当前是不是第 1 件，从而决定按原价还是 6 折计算。

</details>

### 5. 代码填空：第二件商品开始 6 折
