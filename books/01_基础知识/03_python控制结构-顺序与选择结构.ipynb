## 控制结构：程序执行路线的规则

在前面的章节里，我们已经会定义变量，也知道变量有类型。接下来要解决的问题是：这些语句按照什么顺序执行？遇到不同情况时，程序如何做不同选择？

控制结构就是用来描述“代码执行路线”的规则。本章只介绍两类最基础的控制结构：

- 顺序结构：从上到下，一行接一行执行。
- 选择结构：根据条件是否成立，选择执行某一段代码。

顺序结构的一般流程如下：

```{mermaid}
flowchart TD
    A[开始] --> B[语句 1]
    B --> C[语句 2]
    C --> D[语句 3]
    D --> E[结束]
```

选择结构的一般流程如下：

```{mermaid}
flowchart TD
    A[开始] --> B{条件是否成立?}
    B -- 是 --> C[语句 1]
    B -- 否 --> D[语句 2]
    C --> E[结束]
    D --> E[结束]
```

这两张图要观察两点：第一，顺序结构不会跳过中间语句；第二，选择结构会先判断条件，再选择其中一条路径执行。

## 顺序结构

### 数据定义与执行范围

顺序结构是最简单的控制结构：程序从上到下执行代码。上一行得到的结果，可以被下一行继续使用。

例如计算商品总价时，通常需要先有单价和数量，再计算总价，最后输出结果。如果顺序反了，比如还没有定义 `price` 就计算 `total`，程序就无法知道 `price` 是什么。

### 执行流程

顺序结构没有分支，也不会跳过中间步骤。它适合描述“必须一步一步完成”的任务。

```{mermaid}
flowchart TD
    A[定义商品单价 price] --> B[定义商品数量 count]
    B --> C[计算 total = price * count]
    C --> D[输出 total]
```

这个流程中，`total` 依赖 `price` 和 `count`。所以先定义变量，再使用变量，是顺序结构里最基本的规则。

````python
price = 12.5
count = 3

total = price * count

print("商品单价:", price)
print("商品数量:", count)
print("商品总价:", total)
````

### 顺序结构中的变量流动

顺序结构里，变量的值可以一步一步更新。阅读这类代码时，不要只看最后一行，而要沿着代码顺序追踪变量变化。

下面仍然使用商品购买场景：先得到商品小计，再扣除优惠金额，最后得到应付金额。

````python
subtotal = 100
print("商品小计:", subtotal)

discount = 10
print("优惠金额:", discount)

total = subtotal - discount
print("应付金额:", total)
````

## 选择结构：`if` 判断

### 数据定义与适用场景

选择结构用于处理“不同情况走不同路线”的问题。本节继续使用同一个商品购买例子：先计算商品小计，再根据会员等级决定折扣。

会员折扣规则如下：

| 会员等级 | 折扣规则 | 折扣率 |
|---|---|---:|
| 普通 | 无折扣 | `1.00` |
| 银卡 | 95 折 | `0.95` |
| 金卡 | 9 折 | `0.90` |

选择结构的核心是条件表达式。条件表达式的结果只有两种：`True` 或 `False`。

### 执行流程

先看最基础的双分支：如果是普通会员，就不打折；否则先按银卡 95 折处理。

```{mermaid}
flowchart TD
    A[计算 subtotal = price * count] --> B{member_level == 普通?}
    B -- True --> C[discount_rate = 1.00]
    B -- False --> D[discount_rate = 0.95]
    C --> E[total = subtotal * discount_rate]
    D --> E[total = subtotal * discount_rate]
    E --> F[输出应付金额]
```

这里要注意：`if` 后面的条件成立时，只执行 `if` 缩进块；条件不成立时，只执行 `else` 缩进块。

````python
price = 100
count = 2
member_level = "银卡"

subtotal = price * count

if member_level == "普通":
    discount_rate = 1.00
else:
    discount_rate = 0.95

total = subtotal * discount_rate

print("商品小计:", subtotal)
print("会员等级:", member_level)
print("折扣率:", discount_rate)
print("应付金额:", total)
````

### 条件表达式

条件表达式通常由比较运算符组成。下面用 `member_level = "银卡"` 作为例子。

| 表达式 | 含义 | 结果 |
|---|---|---|
| `member_level == "普通"` | 是否普通会员 | `False` |
| `member_level == "银卡"` | 是否银卡会员 | `True` |
| `member_level != "金卡"` | 是否不是金卡会员 | `True` |

注意：判断是否相等使用 `==`，不是 `=`。`=` 是赋值。

````python
member_level = "银卡"

print('member_level == "普通" ->', member_level == "普通")
print('member_level == "银卡" ->', member_level == "银卡")
print('member_level != "金卡" ->', member_level != "金卡")
````

## 多分支选择：`if`、`elif`、`else`

### 数据定义与适用场景

当情况不止两种时，可以使用 `elif`。在会员折扣问题中，会员等级有三种：普通、银卡、金卡。

多分支结构会从上到下检查条件。一旦某个条件成立，就执行对应分支，后面的分支不再检查。

### 执行流程

```{mermaid}
flowchart TD
    A[得到 member_level] --> B{member_level == 金卡?}
    B -- True --> C[discount_rate = 0.90]
    B -- False --> D{member_level == 银卡?}
    D -- True --> E[discount_rate = 0.95]
    D -- False --> F[discount_rate = 1.00]
    C --> G[计算 total]
    E --> G[计算 total]
    F --> G[计算 total]
```

阅读多分支代码时，要特别注意条件顺序。这里先判断金卡，再判断银卡，最后把剩下的情况当作普通会员处理。

````python
price = 100
count = 2
member_level = "金卡"

subtotal = price * count

if member_level == "金卡":
    discount_rate = 0.90
elif member_level == "银卡":
    discount_rate = 0.95
else:
    discount_rate = 1.00

total = subtotal * discount_rate

print("商品小计:", subtotal)
print("会员等级:", member_level)
print("折扣率:", discount_rate)
print("应付金额:", total)
````

### 常见写法对比

| 结构 | 适用场景 | 商品购买示例 |
|---|---|---|
| 单分支 `if` | 只在条件成立时额外做一件事 | 如果是金卡会员，额外显示“享受最高折扣” |
| 双分支 `if ... else` | 两种情况必须选一种 | 普通会员无折扣，非普通会员 95 折 |
| 多分支 `if ... elif ... else` | 多种情况选一种 | 普通无折扣 / 银卡 95 折 / 金卡 9 折 |

无论哪一种结构，缩进都非常重要。属于同一个分支的语句，必须保持相同缩进。

````python
member_level = "金卡"
message = "普通结算"

if member_level == "金卡":
    message = "享受最高折扣"

print("会员等级:", member_level)
print("提示:", message)
````

### 1. 概念判断：为什么代码顺序重要？

下面两行代码能否交换顺序？为什么？

```python
total = price * count
price = 12.5
```

<details>
<summary>参考答案</summary>

不能直接交换。`total = price * count` 需要先知道 `price` 和 `count` 的值。如果在定义 `price` 之前使用它，程序无法完成计算，会出现名称未定义的问题。

</details>

### 2. 概念判断：`=` 和 `==` 有什么区别？

请说明下面两种写法分别表示什么：

```python

member_level == "银卡"
```

<details>
<summary>参考答案</summary>

`member_level = "银卡"` 是赋值，把字符串 `"银卡"` 存入变量 `member_level`。`member_level == "银卡"` 是比较，判断当前会员等级是否为银卡，结果是 `True` 或 `False`。

</details>
