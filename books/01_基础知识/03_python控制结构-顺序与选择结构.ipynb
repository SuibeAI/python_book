## 控制结构：程序执行路线的规则

在前面的章节里，我们已经会定义变量，也知道变量有类型。接下来要解决的问题是：这些语句按照什么顺序执行？遇到不同情况时，程序如何做不同选择？

控制结构就是用来描述“代码执行路线”的规则。本章只介绍两类最基础的控制结构：

- 顺序结构：从上到下，一行接一行执行。
- 选择结构：根据条件是否成立，选择执行某一段代码。

```{mermaid}
flowchart TD
    A[开始] --> B[读取或定义变量]
    B --> C[按顺序计算中间结果]
    C --> D{条件是否成立?}
    D -- 是 --> E[执行 if 分支]
    D -- 否 --> F[执行 else 分支]
    E --> G[输出结果]
    F --> G[输出结果]
```

这张图要观察两点：第一，进入条件判断之前，代码仍然是顺序执行；第二，进入选择结构之后，程序只会走符合条件的那条路径。

## 顺序结构

### 数据定义与执行范围

顺序结构是最简单的控制结构：程序从上到下执行代码。上一行得到的结果，可以被下一行继续使用。

例如计算商品总价时，通常需要先有单价和数量，再计算总价，最后输出结果。如果顺序反了，比如还没有定义 `price` 就计算 `total`，程序就无法知道 `price` 是什么。

### 执行流程

顺序结构没有分支，也不会跳过中间步骤。它适合描述“必须一步一步完成”的任务。

```{mermaid}
flowchart TD
    A[定义商品单价 price] --> B[定义商品数量 count]
    B --> C[计算 total = price * count]
    C --> D[输出 total]
```

这个流程中，`total` 依赖 `price` 和 `count`。所以先定义变量，再使用变量，是顺序结构里最基本的规则。

````python
price = 12.5
count = 3

total = price * count

print("商品单价:", price)
print("商品数量:", count)
print("商品总价:", total)
````

### 顺序结构中的变量流动

顺序结构里，变量的值可以一步一步更新。阅读这类代码时，不要只看最后一行，而要沿着代码顺序追踪变量变化。

````python
score = 80
print("原始成绩:", score)

score = score + 5
print("加上平时分后:", score)

score = score - 2
print("扣除迟交分后:", score)
````

## 选择结构：`if` 判断

### 数据定义与适用场景

选择结构用于处理“不同情况走不同路线”的问题。例如：

- 分数是否及格。
- 年龄是否达到购票优惠条件。
- 温度是否超过预警线。

选择结构的核心是条件表达式。条件表达式的结果只有两种：`True` 或 `False`。

### 执行流程

最基础的双分支结构是 `if ... else ...`。

```{mermaid}
flowchart TD
    A[得到 score] --> B{score >= 60?}
    B -- True --> C[输出 及格]
    B -- False --> D[输出 不及格]
    C --> E[结束]
    D --> E[结束]
```

这里要注意：`if` 后面的条件成立时，只执行 `if` 缩进块；条件不成立时，只执行 `else` 缩进块。

````python
score = 75

if score >= 60:
    result = "及格"
else:
    result = "不及格"

print("成绩:", score)
print("判断结果:", result)
````

### 条件表达式

条件表达式通常由比较运算符组成。下面用 `score = 75` 作为例子。

| 表达式 | 含义 | 结果 |
|---|---|---|
| `score >= 60` | 是否大于等于 60 | `True` |
| `score < 60` | 是否小于 60 | `False` |
| `score == 75` | 是否等于 75 | `True` |
| `score != 75` | 是否不等于 75 | `False` |

注意：判断是否相等使用 `==`，不是 `=`。`=` 是赋值。

````python
score = 75

print("score >= 60 ->", score >= 60)
print("score < 60 ->", score < 60)
print("score == 75 ->", score == 75)
print("score != 75 ->", score != 75)
````

## 多分支选择：`if`、`elif`、`else`

### 数据定义与适用场景

当情况不止两种时，可以使用 `elif`。例如成绩等级通常不是只有“及格/不及格”，还可能分为优秀、良好、合格、需要努力。

多分支结构会从上到下检查条件。一旦某个条件成立，就执行对应分支，后面的分支不再检查。

### 执行流程

```{mermaid}
flowchart TD
    A[得到 score] --> B{score >= 90?}
    B -- True --> C[等级 A]
    B -- False --> D{score >= 80?}
    D -- True --> E[等级 B]
    D -- False --> F{score >= 60?}
    F -- True --> G[等级 C]
    F -- False --> H[需要努力]
```

阅读多分支代码时，要特别注意条件顺序。如果先写 `score >= 60`，那么 95 分也会先进入这个分支，后面的 `score >= 90` 就没有机会执行。

````python
score = 86

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "需要努力"

print("成绩:", score)
print("等级:", grade)
````

### 常见写法对比

| 结构 | 适用场景 | 示例 |
|---|---|---|
| 单分支 `if` | 只在条件成立时额外做一件事 | 如果温度超过 35 度，输出高温提醒 |
| 双分支 `if ... else` | 两种情况必须选一种 | 及格 / 不及格 |
| 多分支 `if ... elif ... else` | 多种情况选一种 | A / B / C / 需要努力 |

无论哪一种结构，缩进都非常重要。属于同一个分支的语句，必须保持相同缩进。

````python
temperature = 38
warning = "正常"

if temperature >= 35:
    warning = "高温提醒"

print("温度:", temperature)
print("提醒:", warning)
````

## 习题

下面的题目用于检查你是否理解顺序结构和选择结构。代码题采用填空形式，填完后运行单元格，若没有报错并输出“验证通过”，说明答案正确。

### 1. 概念判断：为什么代码顺序重要？

下面两行代码能否交换顺序？为什么？

```python
total = price * count
price = 12.5
```

<details>
<summary>参考答案</summary>

不能直接交换。`total = price * count` 需要先知道 `price` 和 `count` 的值。如果在定义 `price` 之前使用它，程序无法完成计算，会出现名称未定义的问题。

</details>

### 2. 概念判断：`=` 和 `==` 有什么区别？

请说明下面两种写法分别表示什么：

```python
score = 75
score == 75
```

<details>
<summary>参考答案</summary>

`score = 75` 是赋值，把 75 存入变量 `score`。`score == 75` 是比较，判断 `score` 的值是否等于 75，结果是 `True` 或 `False`。

</details>

### 3. 概念判断：多分支为什么要注意顺序？

如果判断成绩等级时，先写 `score >= 60`，再写 `score >= 90`，会出现什么问题？

<details>
<summary>参考答案</summary>

如果先判断 `score >= 60`，那么 90 分以上的成绩也满足这个条件，会先进入 60 分对应的分支。后面的 `score >= 90` 就不会再执行，等级判断会出错。

</details>

### 4. 代码填空：顺序结构计算总价

请补全代码，先计算小计，再计算优惠后的总价。

````python
price = 20
count = 3
discount = 5

# 填空 1：计算小计
subtotal = "请填空"

# 填空 2：计算优惠后的总价
total = "请填空"

if subtotal == "请填空" or total == "请填空":
    print("请先完成填空。")
else:
    assert subtotal == 60
    assert total == 55
    print("第 4 题验证通过")
````

<details>
<summary>第 4 题参考答案</summary>

```python
subtotal = price * count
total = subtotal - discount
```

</details>

### 5. 代码填空：双分支判断是否及格

请补全条件，让 58 分得到“不及格”。

````python
score = 58

# 填空：写出及格条件
condition = "请填空"

if condition == "请填空":
    print("请先完成填空。")
else:
    if condition:
        result = "及格"
    else:
        result = "不及格"

    assert result == "不及格"
    print("第 5 题验证通过")
````

<details>
<summary>第 5 题参考答案</summary>

```python
condition = score >= 60
```

</details>

### 6. 代码填空：多分支判断等级

请补全两个条件，让 `score = 86` 得到等级 `B`。

````python
score = 86

condition_a = "请填空 1"
condition_b = "请填空 2"

if condition_a == "请填空 1" or condition_b == "请填空 2":
    print("请先完成填空。")
else:
    if condition_a:
        grade = "A"
    elif condition_b:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "需要努力"

    assert grade == "B"
    print("第 6 题验证通过")
````

<details>
<summary>第 6 题参考答案</summary>

```python
condition_a = score >= 90
condition_b = score >= 80
```

</details>
